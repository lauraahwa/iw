```python
import spacy
import spacy.lang.zh.examples

import os
import glob
import pandas as pd
import xml.etree.ElementTree as ET
```

```python
def xml_to_dataframe(xml_file):
    # Parse the XML file
    tree = ET.parse(xml_file)
    root = tree.getroot()
    
    # Create lists to store the data
    sentences = []
    # sentence_ids = []
    
    # Find all 's' elements (sentences) in the XML
    for s_elem in root.findall('.//s'):
        sentence_id = s_elem.get('id')  # Get the ID attribute
        text = s_elem.text  # Get the text content
        
        # sentence_ids.append(sentence_id)
        sentences.append(text)
    
    # Create DataFrame
    # df = pd.DataFrame({
    #     'sentence_id': sentence_ids,
    #     'text': sentences
    # })
    
    return sentences
```

```python
file_list_zh = ['dev/intertext/dev-anno.001_zh.xml', 'dev/intertext/dev-anno.002_zh.xml', 'dev/intertext/dev-anno.003_zh.xml', 'dev/intertext/dev-anno.004_zh.xml', 'dev/intertext/dev-anno.005_zh.xml', 'dev/intertext/dev-anno.006_zh.xml', 'dev/intertext_01/dev-anno-1.001_zh.xml', 'dev/intertext_01/dev-anno-1.002_zh.xml', 'dev/intertext_01/dev-anno-1.003_zh.xml', 'dev/intertext_01/dev-anno-1.004_zh.xml', 'dev/intertext_01/dev-anno-1.005_zh.xml', 'dev/intertext_01/dev-anno-1.006_zh.xml', 'dev/intertext_02/dev-anno-2.001_zh.xml', 'dev/intertext_02/dev-anno-2.002_zh.xml', 'dev/intertext_02/dev-anno-2.003_zh.xml', 'dev/intertext_02/dev-anno-2.004_zh.xml', 'dev/intertext_02/dev-anno-2.005_zh.xml', 'dev/intertext_02/dev-anno-2.006_zh.xml']

cumulative_zh = []
for xml_file in file_list_zh:
    sentences = xml_to_dataframe(xml_file)
    cumulative_zh.append(sentences)

# Flatten the list of lists into a single list
all_sentences_zh = [sentence for sublist in cumulative_zh for sentence in sublist]
```

```python
file_list_en = ['dev/intertext/dev-anno.001_en.xml', 'dev/intertext/dev-anno.002_en.xml', 'dev/intertext/dev-anno.003_en.xml', 'dev/intertext/dev-anno.004_en.xml', 'dev/intertext/dev-anno.005_en.xml', 'dev/intertext/dev-anno.006_en.xml', 'dev/intertext_01/dev-anno-1.001_en.xml', 'dev/intertext_01/dev-anno-1.002_en.xml', 'dev/intertext_01/dev-anno-1.003_en.xml', 'dev/intertext_01/dev-anno-1.004_en.xml', 'dev/intertext_01/dev-anno-1.005_en.xml', 'dev/intertext_01/dev-anno-1.006_en.xml', 'dev/intertext_02/dev-anno-2.001_en.xml', 'dev/intertext_02/dev-anno-2.002_en.xml', 'dev/intertext_02/dev-anno-2.003_en.xml', 'dev/intertext_02/dev-anno-2.004_en.xml', 'dev/intertext_02/dev-anno-2.005_en.xml', 'dev/intertext_02/dev-anno-2.006_en.xml']

cumulative_en = []
for xml_file in file_list_en:
    sentences = xml_to_dataframe(xml_file)
    cumulative_en.append(sentences)

# Flatten the list of lists into a single list
all_sentences_en = [sentence for sublist in cumulative_en for sentence in sublist]
```

```python
all_sentences_en
print(len(all_sentences_en))
```

```python
all_sentences_en

df = pd.DataFrame(all_sentences_en, columns=['sentence'])
df['length'] = df['sentence'].apply(lambda x: len(x.split()))

# filter sentences with lengths between 4 and 30 inclusive
filtered_df = df[(df['length'] >= 6) & (df['length'] <= 30)]

# randomly sample 2117 sentences
sample_size = min(len(filtered_df), 2117)
random_sample = filtered_df.sample(n=sample_size, random_state=42)

# # extract the sentences back to a list if needed
sampled_sentences = random_sample['sentence'].tolist()

# # display sampled sentences
sampled_sentences

random_sample[['sentence']].to_csv('mac_en_sentences.csv', index=False)

print("Sampled sentences have been saved to 'mac_en_sentences.csv'.")
```
